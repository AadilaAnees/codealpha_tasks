```python
df1_enriched['longitude'] = (
    df1_enriched.groupby('Region')['longitude']
    .transform(lambda x: x.fillna(x.dropna().iloc[0] if not x.dropna().empty else np.nan))
)

df1_enriched['latitude'] = (
    df1_enriched.groupby('Region')['latitude']
    .transform(lambda x: x.fillna(x.dropna().iloc[0] if not x.dropna().empty else np.nan))
)
```

```python
avg_unemployment = (
    df1_enriched
    .groupby(['Date', 'Region'])['Estimated Unemployment Rate (%)']
    .mean()
    .reset_index()
)
```

```python
import matplotlib.pyplot as plt

selected_regions = ['Delhi', 'Andhra Pradesh', 'Tamil Nadu', 'Assam']

plot_data = avg_unemployment[
    avg_unemployment['Region'].isin(selected_regions)
]

plt.figure(figsize=(12,6))

for region in selected_regions:
    region_data = plot_data[plot_data['Region'] == region]
    plt.plot(
        region_data['Date'],
        region_data['Estimated Unemployment Rate (%)'],
        label=region
    )

plt.title('Average Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.show()
```

```python
import seaborn as sns

heatmap_data = (
    df1_enriched
    .pivot_table(
        values='Estimated Unemployment Rate (%)',
        index='Region',
        columns='Date',
        aggfunc='mean'
    )
)

plt.figure(figsize=(14,10))
sns.heatmap(heatmap_data, cmap='YlOrRd')
plt.title('Unemployment Rate by Region Over Time')
plt.show()
```

```python
region_avg = (
    df1_enriched
    .groupby(['Region', 'latitude', 'longitude'])
    ['Estimated Unemployment Rate (%)']
    .mean()
    .reset_index()
)
```

# Summary of Key Findings on Unemployment in India:

**Overall Unemployment Trend:** The Estimated Unemployment Rate (%) across India shows significant fluctuations over time, with a notable increase in mid-2020(covid-19 period), likely attributable to the global economic impact during that period. The average unemployment rate across all regions is around 11.79%, with a high standard deviation, indicating considerable variability.


**Regional Disparities**: There are substantial regional differences in unemployment rates. The bar chart (Average Unemployment Rate by Region) clearly shows that states like Tripura, Haryana, and Jharkhand have significantly higher average unemployment rates, while regions like Meghalaya, Odisha, and Assam tend to have lower rates. This highlights the uneven distribution of unemployment challenges across the country.


**Temporal Patterns in Selected Regions:** The line plot (Average Unemployment Rate Over Time) for selected regions (Delhi, Andhra Pradesh, Tamil Nadu, Assam) illustrates diverse temporal patterns. Some regions experienced sharper spikes in unemployment during certain periods (e.g., Delhi and Andhra Pradesh in mid-2020), while others showed more stable or gradually changing rates (e.g., Assam).


**Correlation Analysis:** The correlation matrix reveals the following relationships with the 'Estimated Unemployment Rate (%)':


**Estimated Employed:** There is a weak negative correlation (-0.22) between 'Estimated Employed' and 'Estimated Unemployment Rate (%)'. This is somewhat counter-intuitive, as higher employment would typically lead to lower unemployment. This weak negative correlation might suggest that an increase in the employed population doesn't always translate directly into a proportional decrease in the unemployment rate, possibly due to a growing labor force or other complex economic factors.
Estimated Labour Participation Rate (%): There is almost no correlation (0.002) between 'Estimated Labour Participation Rate (%)' and 'Estimated Unemployment Rate (%)'. This suggests that changes in the percentage of the population actively participating in the labor force do not directly or strongly influence the overall unemployment rate.


**Area-wise Unemployment (Rural vs. Urban):** The bar plot 'Average Unemployment by Area' indicates that urban areas generally experience a higher average unemployment rate compared to rural areas. (The kernel state shows rural average at 10.32% and urban at 13.17%). This could be due to various factors such as differing economic structures, migration patterns, and access to employment opportunities in these distinct areas.
